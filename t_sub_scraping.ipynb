```python
%reset -fs
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time, os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import numpy as np
```

```python
df = pd.read_csv('music_all_90.csv')
df = df.loc[df['language'] != 'Korean']
df = df.drop(['rank', 'logo'], axis = 1)
df['start'] = np.nan
df.head()
```

```python
for i in df['channel'].head():
    print(i)
```

```python
chromedriver = "/Applications/chromedriver" # initiating selenium
os.environ["webdriver.chrome.driver"] = chromedriver
driver = webdriver.Chrome(chromedriver)
web_address = 'https://twitchstats.net/streamer/'
subs = []
start = []
names = []
```

```python
# This will break, but it will get all necessary data first
for i in df['channel']:
    driver.get(web_address + i)
    time.sleep(3 + np.random.rand())
    try:
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        # reading in the html
    except: # In case the page doesn't load properly
        names.append(i)
        subs.append('n/a')
        start.append('n/a')
        driver.close()
        driver = webdriver.Chrome(chromedriver)
    else: # Scrape name, subs, and start date
        names.append(soup.find('div', {'class': 'mst strmst'}).find('h1').text)
        subs.append(soup.find('div', {'class':'bx4'}).find('div' , {'class':'abxnm1'}).text) # reading in subs
        start.append(soup.find('div', {'class':'bx3'}).find('div' , {'class':'abxnm4'}).findNext('div').find('div', {'class':'abxnm4'}).text) # reading in start date
```

```python
df_subs = pd.DataFrame({'name':names, 'subs':subs, 'start_date':start})
df_subs.sample(40)
```

```python

df_subs.sample(40)
```

```python
df_subs = df_subs.loc[df_subs['subs'] != 'n/a'] # Getting rid of 'n/a' and '-'
df_subs = df_subs.loc[df_subs['subs'] != '-']
```

```python
def strstrip(x):  # Formatting str -> int
    x = x.replace(",","")
    return int(x.strip("+"))
```

```python
df_subs['subs'] = df_subs['subs'].apply(strstrip)
df_subs.head()
```

```python
df_subs.to_csv("music_subs_v2.csv")
```
